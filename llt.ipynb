```python
mersenne_primes = [6, 28, 496, 8128, 33550336, 8589869056, 137438691328]
```

```python
def llt(p, found):
    s = 4
    M = 2**p - 1

    for index in range(p - 2):
        print(f"evaluating {p} as candidate for perfect number - "+str(round(index/(p-2) * 100, 2))+ f'% completed')
        s = (s * s - 2) % M
        for f in found:
            print(f"{f} found to be a prime")
        
        print('\n')



    return s == 0
```

```python
found = []
for m in mersenne_primes:
    found.append(m)
    print(f"{m} is {'prime' if llt(m, found) else 'composite'}")
```
